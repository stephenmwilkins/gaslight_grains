```python
import numpy as np
from unyt import cm, pc
from synthesizer.photoionisation import cloudy23 as cloudy
import matplotlib.pyplot as plt
from astropy.io import ascii
import cmasher as cmr

# set style
plt.style.use('../../../matplotlibrc.txt')
```

### Nebular continuum

```python


model = 'default'
index = 1


for index in np.arange(1,11):

    # read the cloudy output continuum file containing the spectra
    spec_dict = cloudy.read_continuum(
        f'{model}/{index}',
        return_dict=True)
    
    

    plt.loglog(spec_dict['lam'], spec_dict['nebular_continuum'], label=index)

plt.legend()
plt.xlim([100, 100000])
plt.show()
```

### Transmission

```python

for index in np.arange(1,11):

    # read the cloudy output continuum file containing the spectra
    spec_dict = cloudy.read_continuum(
        f'{model}/{index}',
        return_dict=True)
    
    transmission = spec_dict['transmitted']/spec_dict['incident']

    plt.loglog(spec_dict['lam'], transmission, label=index)

plt.legend()
plt.xlim([10, 100000])
plt.show()
```
